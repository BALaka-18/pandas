# Q.8. Write a function that returns a team v/s team record.

```python
match
```

```python
def func(t1,t2):
    mask1 = match['team1'] == t1
    mask2 = match['team2'] == t2
    mask3 = match['team1'] == t2
    mask4 = match['team2'] == t1

    print(match[(mask1 & mask2) | (mask3 & mask4)]['winner'].value_counts())
    
func("Kolkata Knight Riders", "Royal Challengers Bangalore")
```
